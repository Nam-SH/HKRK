```python
#!/bin/python3

import os
import sys
from collections import deque


class Node:
    def __init__(self, d):
        self.data = d


def tree(indexes):
    children = [list(map(lambda x: None if x == -1 else Node(x), x)) for x in indexes]
    nodes = {n.data: n for n in filter(None, sum(children, []))}
    nodes[1] = Node(1)
    for idx, val in enumerate(children):
        nodes[idx + 1].left, nodes[idx + 1].right = val
    return nodes[1]


def inOrder(root):
    lst = []
    curr = root
    
    res = []
    while lst or curr:
        if curr:
            lst += [curr]
            curr = curr.left
        elif lst:
            curr = lst.pop()
            res += [curr.data]
            curr = curr.right
    return res


def swapNodes(indexes, queries):
    ans = []
    root = tree(indexes)
    for i in queries:
        h = 1
        q = deque()
        q.append(root)
        while q:
            for _ in range(len(q)):
                node = q.popleft()
                if h % i == 0:
                    node.left, node.right = node.right, node.left
                q += filter(None, (node.left, node.right))
            h += 1
        ans += [inOrder(root)]

    return ans


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    indexes = []

    for _ in range(n):
        indexes.append(list(map(int, input().rstrip().split())))

    queries_count = int(input())

    queries = []

    for _ in range(queries_count):
        queries_item = int(input())
        queries.append(queries_item)

    result = swapNodes(indexes, queries)

    fptr.write('\n'.join([' '.join(map(str, x)) for x in result]))
    fptr.write('\n')

    fptr.close()
```
